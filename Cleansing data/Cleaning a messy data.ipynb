```python
listed = []
current_list = []

for i in content:
    if i == '':
        if current_list:  # Check if current_list is not empty before appending
            listed.append(current_list)
        current_list = []  # Reset current_list for the next sublist
    else:
        current_list.append(i)

# Append the last list if it exists
if current_list:
    listed.append(current_list)
```

```python
data={'word':[],'prononunce':[],'Type':[],'Meaning':[],'Example':[],'Example2':[],'others':[]}
data1=['word', 'prononunce', 'Type', 'Meaning', 'Example','Example2','others']
for i in listed:
    for j in range(len(i)):
        data[data1[j]].append(i[j])

for i in data1:
    if len(data[i])!=75:
        total=75-len(data[i])
        for j in range(total):
            data[i].append('Nan')
df=pd.DataFrame(data)
```
